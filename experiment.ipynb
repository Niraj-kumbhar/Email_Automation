```python
from email.mime.application import MIMEApplication
from email.policy import default
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
import pandas as pd
import datetime
```

```python
security = {}
#Slog.write(f"{datetime.now()}-->Reading security file\n")
with open('security/security.txt','r') as f:
    for line in f:
        line = line.strip()
        name, var=line.partition("=")[::2]
        security[name.strip()] = str(var).strip()
# set security
sender = security['Sender_email'] # change if sender file name is different
pass_key = security['AppPasscode']

body = open('email_config/body.txt','r').read() 

to = ['kumbharniraj1@gmail.com','spidersniff2324@gmail.com']
```

```python
class MailTest:
    def __init__(self, sender, pass_key) -> None:
        #sender = self.sender
        self.s = smtplib.SMTP('smtp.gmail.com', 587)
        self.s.starttls()
        self.s.login(sender, pass_key) # Authentication of sender, generated by app password

    def mail(self, toadd):
        msg = MIMEMultipart() 
        msg['From'] = sender 
        msg['To'] =  toadd
        msg['Subject'] = 'Test 1.0 class'
        msg.attach(MIMEText(body, 'html'))
        text = msg.as_string()
        self.s.sendmail(sender, toadd, text)

    def quitCon(self):
        self.s.quit()
    
```

```python
start = datetime.datetime.now()
mail1 = MailTest(sender=sender, pass_key=pass_key)
try:
    for toadd in to:
        mail1.mail(toadd)

except Exception as e:
    print(e)

finally:
    mail1.quitCon()
    print(datetime.datetime.now()-start)
```
